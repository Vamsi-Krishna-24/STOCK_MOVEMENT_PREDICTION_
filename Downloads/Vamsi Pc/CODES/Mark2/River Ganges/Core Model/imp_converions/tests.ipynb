```python
#defining datasets & CONVERSION coco to labels

import os
import json

# Dataset locations
datasets = {
    "train": {
        "images": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/train/images",
        "coco_file": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/train/_annotations.coco.json",
        "labels": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/train/labels",
    },
    "test": {
        "images": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/test/images",
        "coco_file": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/test/_annotations.coco.json",
        "labels": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/test/labels",
    },
    "valid": {
        "images": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/valid/images",
        "coco_file": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/valid/_annotations.coco.json",
        "labels": "/Users/surisettivamsikrishna/Downloads/Vamsi Pc/CODES/Aerial_River_Plastic_Wastes.v5i.coco-segmentation 8.16.57 AM/valid/labels",
    },
}

# Ensure label directories exist
for dataset in datasets.values():
    os.makedirs(dataset["labels"], exist_ok=True)

# Convert COCO annotations to YOLO format
def coco_to_yolo(annotation, img_width, img_height):
    x, y, w, h = annotation["bbox"]
    # Normalize coordinates for YOLO format
    x_center = (x + w / 2) / img_width
    y_center = (y + h / 2) / img_height
    width = w / img_width
    height = h / img_height
    return annotation["category_id"], x_center, y_center, width, height

# Process each dataset
for dataset_name, dataset in datasets.items():
    print(f"Processing {dataset_name} dataset...")

    # Load COCO JSON file
    with open(dataset["coco_file"], "r") as f:
        coco_data = json.load(f)

    # Map image IDs to file names and sizes
    images_info = {img["id"]: (img["file_name"], img["width"], img["height"]) for img in coco_data["images"]}

    # Group annotations by image_id
    annotations_by_image = {}
    for annotation in coco_data["annotations"]:
        image_id = annotation["image_id"]
        if image_id not in annotations_by_image:
            annotations_by_image[image_id] = []
        annotations_by_image[image_id].append(annotation)

    # Generate label files
    for image_id, (file_name, img_width, img_height) in images_info.items():
        image_path = os.path.join(dataset["images"], file_name)
        label_path = os.path.join(dataset["labels"], os.path.splitext(file_name)[0] + ".txt")

        # Skip if the image is not in the folder
        if not os.path.exists(image_path):
            continue

        # Get annotations for this image
        annotations = annotations_by_image.get(image_id, [])

        # Write YOLO labels to file
        with open(label_path, "w") as label_file:
            for annotation in annotations:
                category_id, x_center, y_center, width, height = coco_to_yolo(annotation, img_width, img_height)
                label_file.write(f"{category_id} {x_center} {y_center} {width} {height}\n")

    print(f"Labels generated for {dataset_name} dataset!")

print("All datasets processed successfully!")
```
